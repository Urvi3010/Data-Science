Download the Iris flower dataset or any other dataset into a DataFrame. (e.g.,
https://archive.ics.uci.edu/ml/datasets/Iris ). Scan the dataset and give the inference as:
1. List down the features and their types (e.g., numeric, nominal) available in the dataset.
2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
3. Create a boxplot for each feature in the dataset.
4. Compare distributions and identify outliers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')
```

```python
# 1. List down the features and their types
print("1. Feature Types:")
print(df.dtypes)
print("\nSummary:\nNumeric: sepal_length, sepal_width, petal_length, petal_width\nNominal: species\n")
```

```python
# 2. Histograms for each numeric feature
print("\n2. Histograms:")
df.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle("Histograms of Iris Dataset Features")
plt.tight_layout()
plt.show()
```

```python
# 3. Boxplots for each numeric feature
print("\n3. Boxplots:") 
for column in df.columns[:-1]:  # Skip 'species'
    sns.boxplot(x='species', y=column, data=df)
    plt.title(f"Boxplot of {column} by Species")
    plt.show()
```

```python

# 4. Compare distributions and identify outliers
print("\n4. Inference:")
print("- Sepal Width shows a few outliers, especially in the setosa species.")
print("- Petal features (length and width) are good for distinguishing species.")
print("- Setosa is clearly separable based on petal length/width, while versicolor and virginica overlap slightly.")
```
