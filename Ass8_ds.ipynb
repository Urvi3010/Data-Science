Data Visualization I
1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information about
the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to see if we
can find any patterns in the data.
2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger
is distributed by plotting a histogram.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset('titanic')
```

```python
dataset.head()
```

```python
sns.distplot(x = dataset['age'],bins =10)
```

```python
sns.distplot(x = dataset['age'],bins =10, kde = False)
```

```python
sns.jointplot(x='age',y='fare', data = dataset, kind = 'scatter')
```

```python
sns.jointplot(x='age',y='fare', data = dataset, kind = 'hex')
```

```python
sns.boxplot(x='sex',y='age',data = dataset)
```

```python
sns.boxplot(x='sex',y='age',data = dataset, hue = 'survived')
```

```python
sns.violinplot(x='sex',y='age',data = dataset)
```

```python
sns.violinplot(x='sex',y='age',data = dataset, hue = 'survived')
```

```python
sns.stripplot(x='sex',y='age',data=dataset)
```

```python
sns.stripplot(x='sex',y='age',data=dataset, hue = 'survived')
```

```python
sns.swarmplot(x='sex',y='age',data=dataset)
```

```python
sns.swarmplot(x='sex',y='age',data=dataset, hue = 'survived')
```

```python
sns.barplot(x='sex',y='age',data=dataset)
```

```python
sns.barplot(x='sex',y='age',data=dataset,hue= 'survived')
```

```python
sns.countplot(x='sex',data=dataset)
```

```python
sns.rugplot(x='fare', data=dataset)
```

```python
cor = dataset.corr(numeric_only = True)
```

```python
sns.heatmap(cor)
```

```python
sns.heatmap(cor,annot = True,cmap = "BuPu")
```
